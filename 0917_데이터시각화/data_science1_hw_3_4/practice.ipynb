```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# 데이터 불러오기
file_path = 'data/statistical_data.csv'
df = pd.read_csv(file_path)
```

```python
# 데이터 확인
print('# 상위 5개 행 출력')
print(df.head())  # 상위 5개 행 출력
print()

print('# 데이터 타입 및 결측치 확인')
print(df.info())  # 데이터 타입 및 결측치 확인
print()

print('# 기초 통계 확인')
print(df.describe())  # 기초 통계 확인
```

```python
# 1. 변수 간 관계를 산점도로 시각화 ###
plt.figure(figsize = (8,6))

sns.scatterplot(x=df['변수1'],y=df['변수2'], hue=df['Group'], palette='deep', alpha=0.7)
#hue: 그룹별로 색깔을 다르게 표시한다 - 여기서는 df속 Group별로 다르게 표현한다. 

plt.title("변수1과 변수2 간 관계 (산점도)")
plt.xlabel('변수1')
plt.ylabel('변수2')
plt.legend(title="Group")
plt.show()

# 해석:
# - 변수1과 변수2 간의 관계를 확인
# - 그룹별로 다른 색을 사용하여 분포를 구분
# - 상관 관계가 있는지 시각적으로 확인 가능
```

```python
correlation_matrix = df[['변수1', '변수2']].corr()

plt.figure( figsize = (6,5) )
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('변수 간 상관 행렬')
plt.show()
```

```python
sns.pairplot(df, hue="Group", diag_kind="hist")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
sns.boxplot(x='Group', y='변수2', data = df, palette='pastel')
plt.title('그룹별 변수1 분포')

plt.subplot(1,2,2)
sns.boxplot(x='Group', y='변수2', data=df, palette = "pastel")
plt.title("그룹별 변수2 분포")

plt.tight_layout()
plt.show()
```
