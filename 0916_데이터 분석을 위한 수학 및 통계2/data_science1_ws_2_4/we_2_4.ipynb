```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#특정 교차로에서 1시간동안 도착하는 차량 수 
df = pd.read_csv('data/data.csv')

poisson_samples = df['poisson_samples']
```

```python
poisson_samples
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(poisson_samples, bins=30, kde=True, edgecolor='black')  # [빈칸] 포아송분포 히스토그램 시각화 코드 작성
plt.title('Data generated with Poisson distribution')
plt.xlabel('Number of vehicles arriving within 1 hour')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 정규분포 근사를 위한 평균과 표준편차 계산
lam = 20
num_samples = 1000

mu = np.mean(poisson_samples)  # 기대값(mean) 계산 코드 작성
sigma = np.sqrt(poisson_samples)  # 표준편차(std) 계산 코드 작성

normal_sample  = np.random.normal(mu, sigma, num_samples)
```

정규분포 근사 확인
- λ가 충분히 클 경우, 포아송분포가 정규분포에 가까워지는 현상을 분석
- 포아송분포의 평균과 표준편차를 이용하여 정규분포를 생성

```python
# 포아송분포와 정규분포 비교
plt.figure(figsize=(8, 5))
sns.histplot(poisson_samples, bins=30, kde=True, color='blue', label='Poisson')  # [빈칸] 포아송분포 시각화 코드 작성
sns.histplot(normal_sample, bins=30, kde=True, color='red', label='Normal', alpha=0.6)  # [빈칸] 정규분포 시각화 코드 작성
plt.title('Comparison of Poisson distribution and normal distribution')
plt.xlabel('Number of vehicles arriving within 1 hour')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
